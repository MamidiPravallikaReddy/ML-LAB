```python
#13(a)
import pandas as pd
data = {
    'Item': ['TV', 'TV', 'TV', 'AC'],
    'Company': ['LG', 'VIDEOCON', 'LG', 'SONY'],
    'Rupees': [12000, 10000, 15000, 14000],
    'USD': [700, 650, 800, 750]
}
df = pd.DataFrame(data)
df
```

```python
#13(b)
new_data = [
    {'Item': 'AC', 'Company': 'SAMSUNG', 'Rupees': 60000, 'USD': 850},
    {'Item': 'TV', 'Company': 'SONY', 'Rupees': 13000, 'USD': 780}
]
df = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)
df
```

```python
#13(c)
lg_tv_max_price = df[(df['Item'] == 'TV') & (df['Company'] == 'LG')]['Rupees'].max()
lg_tv_max_price
```

```python
#13(d)
total_price = df['Rupees'].sum()
print(total_price)
```

```python
#13(e)
sony_median_usd = df[df['Company'] == 'SONY']['USD'].median()
sony_median_usd
```

```python
#13(f)
sorted_df = df.sort_values(by='Rupees')
sorted_df
```

```python
#14(a)
import pandas as pd
import numpy as np
dict = {
    'Name': ['Aparna', 'Pankaj', 'Ram', 'Ramesh', 'Naveen', 'Krrishnav', 'Bhawna'],
    'Degree': ['MBA', 'BCA', 'M.Tech', 'MBA', np.nan, 'BCA', 'MBA'],
    'Score': [90.0, np.nan, 80.0, 98.0, 97.0, 78.0, 89.0]
}

Df = pd.DataFrame(dict)
Df
```

```python
#14(b)
Df.groupby('Degree')['Score'].max()
```

```python
#14(c)
Df.fillna({'Degree': 'Unknown', 'Score': 76}, inplace=True)
Df
```

```python
#14(d)
Df.set_index('Name', inplace=True)
Df
```

```python
#14(e)
Df.groupby(['Name', 'Degree'])['Score'].mean()
```

```python
#14(f)
mba_count = (Df['Degree'] == 'MBA').sum()
print(mba_count)
```

```python
#14(g)
bca_mode = Df[Df['Degree'] == 'BCA']['Score'].mode()[0]
print(bca_mode)
```

```python
#1
import pandas as pd
import numpy as np

# 1) Load auto-mpg.data into a DataFrame autodf
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

column_names = ['mpg', 'cylinders', 'displacement', 'horsepower',
                'weight', 'acceleration', 'model_year', 'origin', 'car_name']

autodf = pd.read_csv(url, delim_whitespace=True, names=column_names, na_values='?')
```

```python
#2
print(autodf.describe(include='all'))
```

```python
#3
print(autodf.head(10))
```

```python
#4
print(autodf.isnull().sum())
```

```python
#4(i)
autodf_fill = autodf.fillna(method='ffill')
print(autodf_fill.isnull().sum())
```

```python
#4(ii)
autodf_drop = autodf.dropna()
print(autodf_drop.isnull().sum())
```

```python
#5
max_mileage_car = autodf.loc[autodf['mpg'].idxmax()]
print(max_mileage_car)
```

```python
#6
print(autodf.groupby('cylinders')['displacement'].mean())
```

```python
#7
avg_cylinders = autodf['cylinders'].mean()
print(avg_cylinders)
```

```python
#8
avg_weight = autodf['weight'].mean()
count_heavy_cars = (autodf['weight'] > avg_weight).sum()
print(count_heavy_cars)
```
